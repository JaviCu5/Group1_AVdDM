# Import data <a class="anchor" id="1"></a>

```python
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("analisi.transparenciacatalunya.cat", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(analisi.transparenciacatalunya.cat,
#                  MyAppToken,
#                  username="user@example.com",
#                  password="AFakePassword")

# First 117011 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
#limit és nombre columnes
results = client.get("q2sg-894k", limit=117011)

# Convert to pandas DataFrame
calls_df = pd.DataFrame.from_records(results)

# show loaded table:
calls_df
```

# Characterisation <a class="anchor" id="2"></a>

```python
#Applying filters to focus only on victims of gender-based violence.
filtre1_df= calls_df[calls_df['motiutrucada']=='Violència masclista']
filtre2_df= filtre1_df[filtre1_df['any']!='2022']
filtre3_df=filtre2_df[filtre2_df['personaquetruca']=='La pròpia usuària']
filtre3_df
```

```python
#List of years from 2013 to 2021
year=np.zeros(9)
for i in range (9):
    year[i]=int(2013+i)
```

## Caller <a class="anchor" id="3"></a>

```python
#Count the number of callers who are the victim herself, family members or professionals (3 items)
truca_df=filtre2_df[['any','personaquetruca']]
truca=truca_df.value_counts().sort_index()
#Convert Pandas Series into array
truca_array=truca.array

#9 years
trucades_propia=np.zeros(9)
trucades_altres=np.zeros(9)
for i in range(9):
    #3 items
    trucades_propia[i]=truca_array[3*i+1]
    trucades_altres[i]=truca_array[3*i]+truca_array[3*i+2]

#Create a new DataFrame with the information above
trucades_df = pd.DataFrame()
trucades_df['year']=year
trucades_df['user-herself']=trucades_propia
trucades_df['others']=trucades_altres
```

```python
#define x and y limits
y = range(0, len(trucades_df))
x_altres = trucades_df['others']
x_propia = trucades_df['user-herself']

#define plot parameters
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(9, 6))

#specify background color and plot title
fig.patch.set_facecolor('xkcd:white')
plt.figtext(.5,.9,'Caller',fontsize=15, ha='center')
    
#define victim herself or others
axes[0].barh(y, x_altres, align='center', color='grey')
axes[0].set(title='Others')
axes[0].set(xlim=[0, 9500])
axes[1].barh(y, x_propia, align='center', color='plum')
axes[1].set(title='User herself')

#adjust grid parameters and specify labels for y-axis
axes[1].grid()
axes[0].set(yticks=y, yticklabels=trucades_df['year'])
axes[0].invert_xaxis()
axes[0].grid()

#display plot
plt.savefig("Persona_que_truca.png")
plt.show()
```

## Age of the victim <a class="anchor" id="4"></a>

```python
#Applying a filter to eliminate irrelevant information
truca_df=filtre3_df[filtre3_df['edat']!='No consta']
#Count the number of victims for each age range
truca_df=truca_df[['any','edat']]
truca=truca_df.value_counts().sort_index()
truca_array=truca.array

#Create a new DataFrame with the information above
llista_18=np.zeros(9)
llista_31=np.zeros(9)
llista_41=np.zeros(9)
llista_51=np.zeros(9)
llista_menys_18=np.zeros(9)
llista_mes_60=np.zeros(9)
for i in range (9):
    suma=truca_array[6*i]+truca_array[6*i+1]+truca_array[6*i+2]+truca_array[6*i+3]+truca_array[6*i+4]+truca_array[6*i+5]
    llista_18[i]=(truca_array[6*i]/suma)*100
    llista_31[i]=(truca_array[6*i+1]/suma)*100
    llista_41[i]=(truca_array[6*i+2]/suma)*100
    llista_51[i]=(truca_array[6*i+3]/suma)*100
    llista_menys_18[i]=(truca_array[6*i+4]/suma)*100
    llista_mes_60[i]=(truca_array[6*i+5]/suma)*100
    
trucades_df = pd.DataFrame()
trucades_df['year']=year
trucades_df['<18']=llista_menys_18
trucades_df['18-30']=llista_18
trucades_df['31-40']=llista_31
trucades_df['41-50']=llista_41
trucades_df['51-60']=llista_51
trucades_df['>60']=llista_mes_60
trucades_df_ordre=trucades_df.sort_index(ascending=False)
```

```python
# create donut plots
startingRadius = 1 + (1* (len(trucades_df)-1))
for index, row in trucades_df_ordre.iterrows():
    scenario = row["year"]
    edat_1 = row["<18"]
    edat_2 = row["18-30"]
    edat_3 = row["31-40"]
    edat_4 = row["41-50"]
    edat_5 = row["51-60"]
    edat_6 = row[">60"]
    textLabel = ''
    #textLabel = scenario
    #print(startingRadius)

    donut_sizes = [edat_1,edat_2,edat_3,edat_4,edat_5,edat_6]

    plt.text(0.01, startingRadius + 0.07, textLabel, horizontalalignment='center', verticalalignment='center')
    plt.pie(donut_sizes, radius=startingRadius, startangle=90, colors=['#CD69C9', '#87CEEB','#71C671','#FFB6C1','#FFF68F','#FF8000'],
            wedgeprops={"edgecolor": "white", 'linewidth': 1})

    startingRadius-=1

# equal ensures pie chart is drawn as a circle (equal aspect ratio)
plt.axis('equal')
name=['<18','18-30','31-40','41-50','51-60','>60']
# create circle and place onto pie chart
circle = plt.Circle(xy=(0, 0), radius=1,facecolor='white')
plt.gca().add_artist(circle)
plt.legend(labels=name,loc=(0.85, 0.1))
plt.savefig("Edat.png")
plt.show()
```

## Education <a class="anchor" id="5"></a>

```python
#Count the number of victims with different academic formation
formacio_df=filtre5_df[['any','edat','formacio']]
formacio=formacio_df.value_counts().sort_index()
formacio_array=formacio.array
formacio
```

```python
#Normalization term
suma=np.zeros(4)
for i in range(4):
    norm=0
    for j in range(4):
        norm=norm+formacio_array[4*i+j]
    suma[i]=norm
suma
```

```python
#Create a new DataFrame
primaris=np.zeros(4)
secundaris=np.zeros(4)
superiors=np.zeros(4)
sense=np.zeros(4)
for i in range(4):
    primaris[i]=formacio_array[4*i]
    secundaris[i]=formacio_array[4*i+1]
    superiors[i]=formacio_array[4*i+2]
    sense[i]=formacio_array[4*i+3]
    
#Normalization
for i,norm in enumerate(suma):
    primaris[i]=(primaris[i]/norm)*100
    secundaris[i]=(secundaris[i]/norm)*100
    superiors[i]=(superiors[i]/norm)*100
    sense[i]=(sense[i]/norm)*100
```

```python
formacio_df = pd.DataFrame()
formacio_df['Primary studies']=primaris
formacio_df['Secondary studies']=secundaris
formacio_df['Higher education']=superiors
formacio_df['No studies']=sense
formacio_df.index=index3
formacio_df
```

```python
index_formacio=['Primary studies','Secondary studies','Higher education','No studies']
```

```python
laboral_df = pd.DataFrame()
laboral_df['Unemployed']=llista1
laboral_df['Student']=llista2
laboral_df['Pensioner']=llista3
laboral_df['Homemaker']=llista4
laboral_df['Employed']=llista5
laboral_df.index=index3
laboral_df
```

```python
#Create a new DataFrame
llista1_dep=np.zeros(6)
llista2_dep=np.zeros(6)
llista3_dep=np.zeros(6)
llista4_dep=np.zeros(6)
llista5_dep=np.zeros(6)
for i in range (6):
    llista1_dep[i]=(dep_array[5*i]/suma_dep[i])*100
    llista2_dep[i]=(dep_array[5*i+1]/suma_dep[i])*100
    llista3_dep[i]=(dep_array[5*i+2]/suma_dep[i])*100
    llista4_dep[i]=(dep_array[5*i+3]/suma_dep[i])*100
    llista5_dep[i]=(dep_array[5*i+4]/suma_dep[i])*100  
```
